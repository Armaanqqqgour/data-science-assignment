### 10. How can you tell the difference between break and continue?

Ans:- Break and continue are statements used within loops (such as for and while loops).

break statement: It is ued to terminate the loop and transfer control to the next statement after the loop
It is typically used to exit a loop prematurely when a certain condition is met.
After the break statement is executed, the remaining iterations of the loop are skipped, and the program continues with the next line of code following the loop.For example:-


```python
for i in range(1,8):
    
    if i==4:
        break
    print(i)    # here in this loop using break statement at i==4 the loop is terminated.
```

continue statement:In loop the continue statement immediately skips the rest of the current iteration and moves to the next iteration of the loop.
It is used to bypass specific iterations of the loop when a certain condition is met, but the loop itself is not terminated.

```python
for i in range(1, 8):
    if i ==5:
        continue
    print(i)       # here in this loop using continue statement at i==5, 5 is skkiped and the iteration or loop continues
```

### 11. In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?

Ans:- In a for loop, the three expressions range(10), range(0, 10), and range(0, 10, 1) are slightly different ways of specifying the range of values to iterate over.

range(10):

This expression generates a sequence of numbers starting from 0 up to 9 and not including, the specified end value, which is 10.
It is equivalent to range(0, 10, 1).its assumes the step value as 1(by default).

range(0, 10):

This expression explicitly specifies the start value as 0 and the end value as 10.
The sequence of numbers generated will start from the specified start value (0) and go up to, but not including, the specified end value (10).its assumes the step value as 1(by default).

range(0, 10, 1):
This expression explicitly specifies the start value as 0, the end value as 10, and the step value as 1.
The sequence of numbers generated will start from the specified start value (0), increment by the specified step value (1), and stop before reaching the specified end value (10).
In this case, the step value of 1 indicates that the generated sequence will increment by 1 for each iteration.

we can say that all the three expression are interchangeable that generate the same sequence


```python
print(range(10))
print(range(0,10))
print(range(0,10,1))
```
